```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    if isinstance(text, str):  # Check if text is a string
        tokens = word_tokenize(text)
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
        clean_text = re.sub(r'[^\w\s]', '', ' '.join(filtered_tokens))
        lemmatizer = WordNetLemmatizer()
        lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(clean_text)]
        return ' '.join(lemmatized_words)
    else:
        return ''

text = "This is a sample sentence with special characters!!! 🎉"
preprocessed_text = preprocess_text(text)
print(preprocessed_text)
```

```python
# data1['tweet'] = data1['tweet'].apply(preprocess_text)
data1['tweet'] = data1['tweet'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(data1['tweet'])
```

```python
from sklearn.preprocessing import LabelEncoder
le_senti = LabelEncoder()
data1['sentiment'] = le_senti.fit_transform(data1['sentiment'])
```

```python
data1.head()
```

```python
X = data1['tweet']
Y = data1['sentiment']
X = vect.transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(lr_pred,y_test)
print(lr_acc)
```

```python
lr_pred
```

```python
from sklearn.svm import LinearSVC
SVCmodel = LinearSVC()
SVCmodel.fit(X_train,y_train)
```

```python
svc_pred = SVCmodel.predict(X_test)
svc_acc = accuracy_score(svc_pred,y_test)
print(svc_acc)
```
